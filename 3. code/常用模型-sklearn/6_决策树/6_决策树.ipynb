# 决策树
+ 应用于**分类、回归、多输出**
+ 可以对复杂的数据集进行拟合
+ 不需要太多的数据预处理，特别是**特征的缩放或者归一化**

# 训练并可视化 

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
x = iris.data[: , 2:]
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
```

```python
# 对其进行可视化
#
from sklearn.tree import export_graphviz
tree_pic = export_graphviz(
    tree_clf,
    out_file="iris_tree.dot",
    feature_names=iris.target_names[1:],
    rounded=True,
    filled=True
)
```

+ samples--统计出他应用于多少个训练样本实例
+ values--每个节点中每个类别有多少样例

# 估计分类概率

```python
#估算某个样例属于某一类别的概率 predict_proba
prob = tree_clf.predict_proba([[5,1.5]])
print(prob)
```
