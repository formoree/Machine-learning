# 线性回归 

+ 针对一个简单的线性回归模型，我们将使用两种不同训练方法得到模型最优解：
     + 封闭方程求根运算
     + 迭代优化：梯度下降

## 正规方程

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = 2*np.random.rand(100,1)
y = 4 + 3*x +np.random.randn(100,1)

#绘制散点图
plt.figure(figsize=(10, 10), dpi=100)
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#np.c_ 用于连接两个矩阵 np.ones（）函数返回给定形状和数据类型的新数组，其中元素的值设置为1
#inv()求矩阵的逆 dot()矩阵乘法
x_b =np.c_[np.ones((100,1)),x] #左右相连--这里为什么会相连？
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y) #求解最小化损失值theta
theta_best
```

```python
#使用theta_best来预测
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)),x_new]
y_predict = x_new_b.dot(theta_best)
y_predict
```

```python
#画出模型图像
plt.plot(x_new,y_predict,"r-")
plt.plot(x,y,'b.')
plt.axis([0,2,0,15])
plt.show()
```

```python
#调用sklearn模型进行运算
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x,y)
print(LR.intercept_,LR.coef_)
print(LR.predict(x_new))
```

## 梯度下降--适用于特征多 训练实例多

+ 梯度下降的两个主要挑战
    + 初始值在图像左侧：容易陷入局部最小值
    + 初始值在图像右侧：训练时间太长 无法到达全局最小值
    
**使用梯度下降时，我们一定要保证所有的特征有着相近的尺度范围，要不然我们会使用较长时间收敛。**

### 批量梯度下降 

```python
eta = 0.1 #学习率
n_iterations = 1000 #迭代次数
m = 100

theta = np.random.rand(2,1)

for interation in range(n_iterations):
    gradients = 2/m*x_b.T.dot(x_b.dot(theta)-y)
    theta = theta-eta*gradients # array([[4.11424371],
                                #           [2.89192183]])
    
```

### 随机梯度下降
+ 训练所有数据集较少
+ 大体上呈下降趋势 下降不规律(正因为其不规律性，它可以跳过局部最小值)
+ 最后参数值不错 但不是最优解(降低学习率可以解决这个问题)
+ 模拟退火：开始时，走的每一步较大，然后变得越来越小，从而到达全局最小值。(类似金属冷却过程)

```python
n_epoches =50
t0,t1 = 5,50 #learning_schedule的超参数

def learning_schedule(t):
    return t0/(t+t1)

for epoch in range(n_epoches):
    for i in range(m):
        random_index = np.random.randint(m) # 随机种子
        #随机拿取小批量数据集
        xi = x_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2*xi.T.dot(xi.dot(theta)-yi)
        #更新eta
        eta = learning_schedule(epoch*m+i)
        #下降
        theta = theta-eta*gradients
theta
```

```python
#调用sklearn内置API
from sklearn.linear_model import SGDRegressor
sgd_r = SGDRegressor(n_iter_no_change=50,penalty=None,eta0=0.1)
sgd_r.fit(x,y)
#sgd_r.fit(x,y.ravel()) # ravel()将数组拉为一维
print(sgd_r.coef_,sgd_r.intercept_)
```

### 小批量梯度下降 Mini-batch GD 
小批量梯度下降在参数空间上的表现比随机梯度下降要好很多，尤其是在有大量的小型实例集。

# 多项式回归 

**当存在多个特征时，多项式回归能够找出特征之间的关系(普通线性回归无法做到)**

```python
#随机初始化数据
m = 100
X = 6*np.random.rand(m,1)-3  #建立(100,1)大小
Y = 0.5*X**2 + X + 2 + np.random.randn(m,1)

#让训练集中的平方项作为新特征
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2,include_bias=False)
X_poly = poly_features.fit_transform(X)

#导入线性回归中运算
lin_reg = LinearRegression()
lin_reg.fit(X_poly,y)
# lin_reg.intercept_,lin_reg.coef_
X_poly[0]
```

## 学习曲线

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors= [], []
    for m in range(1,len(X_train)):
        model.fit(X_train[:m],y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    return X_train, X_val, y_train, y_val

    
lin_reg = LinearRegression()
X_train, X_val, y_train, y_val = plot_learning_curves(lin_reg, X, Y)
```

+ 分析：
    + 前期：模型能较好拟合。但加入新样本后，拟合程度变差。(原因：1.含有噪音 2.数据集非线性)
    + 中后期：模型拟合能力变平稳 **模型欠拟合，增加数据集是没用的，我们需要寻找拥有更多特征的模型**
    
+ 过拟合：训练数据的误差较低，但是曲线之间存在缝隙。
    + 解决方案：使用更大的训练集

```python
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
    ("poly_fetures", PolynomialFeatures(degree=4, include_bias=False)),
    ("lin_reg", LinearRegression()),
])
X_train, X_val, y_train, y_val = plot_learning_curves(polynomial_regression, X, Y)
```

+ 分析：
    + 误差比线性模型低很多
    + 两条线之间有间隔 **模型在训练集上的表现比验证集上的好很多，也是过拟合的显著特征**  
    
+ 一个模型的泛化误差决定因素：
    + 偏差：错误的假设决定
    + 方差：数据的微小变化
    + 不可约误差：数据本身噪声--数据清洗

# 线性模型的正则化
+ 正则化：(限制)，模型越少的自由度，越难拟合数据
+ 对于一个线性模型，正则化的典型就是约束模型中的**参数权重**

## 岭回归 Ridge 

+ 注意：
    + 关于λ(λ决定了正则化模型的强度)：
        + λ=0 岭回归变为线性回归。
        + λ非常的大，所有的权重都接近于0，最终结果是穿过数据平均值的水平直线。
    + 岭回归前，一定是要穿过数据平均值得水平直线。

```python
#岭回归
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X, Y)
ridge_reg.predict([[1.5]])
```

**正确理解：岭回归可以说是一类模型 而非数据处理的某一步**

```python
#随机梯度下降
from sklearn.linear_model import SGDRegressor
#penalty="l2"在损失函数之前加上一项：权重向量的范数平方的一半
sgd_reg = SGDRegressor(penalty="l2")
sgd_reg.fit(X, Y.ravel())
sgd_reg.predict([[1.5]])
```

## lasso回归

+ 特征：倾向于完全消除最不重要的特权的权重(将他们设置为0)

```python
#Lasso回归
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.1)
#等同于lasso_reg = SGDRegressor(penalty="l1")
lasso_reg.fit(X, Y.ravel())
lasso_reg.predict([[1.5]])
```

## 弹性网络 ElasticNet

+ 介于Ridge回归和Lasso回归之间 r为模型混合率(r=0 Ridge回归;r=1 Lasso回归)
+ 三个模型的选择：
    + 正则项表现的较好：**岭回归**
    + 特征有用的较少：**弹性网络和Lasso回归**
    + 特征数量比样本数量大：弹性网络表现的不太好

```python
#弹性网络:结合Lasso和岭回归，是Lasso的改进模式
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, Y)
elastic_net.predict([[1.5]])
```

## 早起停止法(特殊)

一轮一轮的训练之后，最开始RMSE是在下降，但轮次达到一定数量之后会上升，就是过拟合的体现。所以我们规定在误差值达到最小时就停止训练。

# 逻辑回归 
一些回归算法也可以用于分类(反之亦然)

```python
#从sklearn中得到花蕊数据集
from sklearn import datasets
iris = datasets.load_iris()
#打印数据集属性
list(iris.keys())
```

```python
#我们选取花瓣宽度这个属性进行判断
import numpy as np
X = iris["data"][:,3:]
y = (iris["target"] == 2).astype(np.int)

from sklearn.linear_model import LogisticRegression
#训练一个线性回归模型
log_reg = LogisticRegression()
log_reg.fit(X, y)

#画图显示预测结果
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
import matplotlib.pyplot as plt
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris virginica")
```

```python
#预测一下
log_reg.predict([[1.7], [1.5]])
```

## 均方差函数 MSE
+ 主要用于评估回归模型的使用效果(**模型仅仅只有一个全局最小值**)
+ RMSE（Root Mean Square Error）均方根误差 MSE开平方根

## L1范数损失 L1Loss
output 和 target 之差的绝对值

## 交叉熵损失 CrossEntropyLoss 
在多分类任务中，经常采用 softmax 激活函数+交叉熵损失函数

## KL 散度损失 KLDivLoss
计算 input 和 target 之间的 KL 散度。KL 散度可用于衡量不同的连续分布之间的距离, 在连续的输出分布的空间上(离散采样)上进行直接回归时很有效。
