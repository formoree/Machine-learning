# RL的基础知识 

## RL的分类 

+ 另一种分类(根据模型的更新方式)：
    + 回合更新指的是游戏开始后, 等游戏结束之后进行更新。
    + 单步更新则是在游戏进行中每一步都在更新, 不用等待游戏的结束。
+ 其他分类(学习时是否有环境交互)：
    + 离线强化学习。学习过程中，不与环境进行交互，只从dataset中直接学习，而dataset是采用别的策略收集的数据，并且采集数据的策略并不是近似最优策略(Q learning)
    + 在线强化学习。学习过程中，智能体需要和环境进行交互。
        + on-policy采用的是当前策略搜集的数据训练模型，每条数据仅使用一次。Q-learning,DQN,DDPG,SAC,etc.
        + off-policy训练采用的数据不需要是当前策略搜集的。REINFORCE,A3C,PPO,etc.
